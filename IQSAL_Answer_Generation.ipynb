# Bloom's Taxonomy

```python
# Preprocess a single question
def preprocess_question(question):
    question = question.lower()
    question = ''.join([char for char in question if char not in string.punctuation])
    tokens = word_tokenize(question)
    tokens = [token for token in tokens if token not in stopwords.words('english')]
    return tokens
```

```python
bloom_taxonomy_mappings = {
    "Recall or retrieve": ["define", "describe", "identify", "know", "label", "list", "match", "name", "outline", "recall", "recognize", "reproduce", "select", "state", "what"],
    "Understand": ["comprehend", "convert", "defend", "distinguish", "estimate", "explain", "extend", "generalize", "give an example", "infer", "interpret", "paraphrase", "how"],
    "Apply": ["apply", "change", "compute", "construct", "demonstrate", "discover", "manipulate", "modify", "operate", "predict", "why"],
    "Analyze": ["analyze", "break down", "compare", "contrast", "diagram", "deconstruct", "differentiate", "discriminate", "distinguish", "identify", "illustrate"],
    "Evaluate": ["appraise", "compare", "conclude", "contrast", "criticize", "critique", "defend", "describe", "discriminate", "evaluate", "explain", "interpret", "why"],
    "Create": ["categorize", "combine", "compile", "compose", "create", "devise", "design", "explain", "generate", "modify", "organize", "plan", "rearrange"]
}
```

```python
# Retrieve WordNet synsets for Bloom's Taxonomy wordlist
def get_synsets(wordlist):
    synsets = {}
    for word in wordlist:
        synsets[word] = wordnet.synsets(word)
    return synsets
```

```python
# Classify a single question into Bloom’s Taxonomy categories
def classify_question(question, bloom_taxonomy_mappings):
    category_matches = {category: 0 for category in bloom_taxonomy_mappings}
    verbs = preprocess_question(question)
    for verb in verbs:
        for category, verb_list in bloom_taxonomy_mappings.items():
            if verb in verb_list:
                category_matches[category] += 1
    max_category = max(category_matches, key=category_matches.get)
    return max_category
```

```python
# Apply classification to each question in the DataFrame
def classify_questions(df, column_name):
    classifications = []
    for question in df[column_name]:
        classification = classify_question(question, bloom_taxonomy_mappings)
        classifications.append(classification)
    return classifications
```

```python
df['BT Classification'] = classify_questions(df, 'Question')
```

```python
def num_BT(df, column_name):
  BT = []
  for item in df[column_name]:
    if item == "Recall or retrieve":
      BT.append(1)
    elif item == "Understand":
      BT.append(2)
    elif item == "Apply":
      BT.append(3)
    elif item == "Analyze":
      BT.append(4)
    elif item == "Evaluate":
      BT.append(5)
    elif item == "Create":
      BT.append(6)
  return BT
```

```python
df['BT'] = num_BT(df, 'BT Classification')
df.head()
```

```python
# Define a function to generate a question for each row
def generate_Q1(row):
    # Extract the question and answer from the row
    question = row['Question']
    answer = row['Mark Scheme']

    # Define your prompt template dynamically for each question
    prompt_template = PromptTemplate(
        template=f'''Assume you have no knowledge of external information other than what is provided as input. Given this a scenario-based question: {question},
        generate a focused question that directly addresses the task without adding unrelated information. Ensure that the newly generated question is in the same
        cognitive difficulty level of the original question as according to Bloom's Taxonomy by preserving the Bloom's Taxonomy verb choice in the question.
        The question must not contain any specific scenario-based information or any examples. This question that you create, must be so generalized, that all
        similar scenario-based questions can also be condensed into this basic generated question.
        For context on what is the key information that the question is supposed to answer, this is the mark scheme: {answer}
        Only output one single question. Only use information provided in the question or the mark scheme.
        '''
    ,
        input_variables=['question', 'answer'] # Specify the input variables here
    )

    # Create the LLMChain with your LLM and prompt template
    chain = LLMChain(llm=llm, prompt=prompt_template)

    # Run the chain with your input
    response = chain.run({
        'question': question,
        'answer': answer
    })

    return response
```

# Jaccard Similarity and Further Question Generation

```python
def preprocess_text(text):
    stop_words = set(stopwords.words('english'))
    punctuation = string.punctuation + "“”"  # Add additional punctuation marks if necessary
    tokens = word_tokenize(text)
    # Remove punctuation and stopwords, and convert to lowercase
    filtered_tokens = [word.lower() for word in tokens if word.lower() not in stop_words and word not in punctuation]
    return set(filtered_tokens)  # Convert to set for Jaccard similarity calculation
```

```python
def gen_question_iteration(df, threshold):
    jaccard_similarities_1  = df.apply(lambda row: len(preprocess_text(row['GQ_Q2']).intersection(preprocess_text(row['GQ2']))) / len(preprocess_text(row['GQ_Q2']).union(preprocess_text(row['GQ2']))), axis=1)
    jaccard_similarities_2  = df.apply(lambda row: len(preprocess_text(row['GQ_MS2']).intersection(preprocess_text(row['GQ2']))) / len(preprocess_text(row['GQ_MS2']).union(preprocess_text(row['GQ2']))), axis=1)
    average_similarity_1 = jaccard_similarities_1 .mean()
    average_similarity_2 = jaccard_similarities_2 .mean()
    print("Similarity Difference: ", abs(average_similarity_1 - average_similarity_2))

    while abs(average_similarity_1 - average_similarity_2) > threshold:
        df['GQ_Q2'] = df.apply(generate_GQ, axis=1, q1_col='GQ_Q2', q2_col='GQ2')
        df['GQ_MS2'] = df.apply(generate_GQ, axis=1, q1_col='GQ_MS2', q2_col='GQ2')
        df['GQ2'] = df.apply(generate_GQ, axis=1, q1_col='GQ_Q2', q2_col='GQ_MS2')

        jaccard_similarities_1  = df.apply(lambda row: len(preprocess_text(row['GQ_Q2']).intersection(preprocess_text(row['GQ2']))) / len(preprocess_text(row['GQ_Q2']).union(preprocess_text(row['GQ2']))), axis=1)
        jaccard_similarities_1  = df.apply(lambda row: len(preprocess_text(row['GQ_MS2']).intersection(preprocess_text(row['GQ2']))) / len(preprocess_text(row['GQ_MS2']).union(preprocess_text(row['GQ2']))), axis=1)
        average_similarity_1 = jaccard_similarities_1 .mean()
        average_similarity_2 = jaccard_similarities_1 .mean()
        print("Similarity Difference: ", abs(average_similarity_1 - average_similarity_2))

    return df
```

```python
model_name = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)
```

```python
# Tokenize and encode sentences in the DataFrame column
def encode_sentences(sentence):
    inputs = tokenizer(sentence, return_tensors="pt", padding=True, truncation=True)
    with torch.no_grad():
        outputs = model(**inputs)
        embeddings = outputs.last_hidden_state.mean(dim=1)  # Average pooling over tokens to get sentence embeddings
    return embeddings
```

# Advanced Exam Question Similarity

```python
selected_questions = set()
def bloom_similarity(question1_index, question2_index):
    global selected_questions
    # Ensure question1_index and question2_index are integers
    if isinstance(question1_index, int) and isinstance(question2_index, int):
        if question1_index in selected_questions and question2_index in selected_questions:
            return False
        else:
            question1 = df.loc[question1_index]
            question2 = df.loc[question2_index]
            return question1['BT'] == question2['BT']
    else:
        raise TypeError("Indices must be integers")

def semantic_similarity(question1_index, question2_index):
    global selected_questions
    if question1_index in selected_questions and question2_index in selected_questions:
        return 0
    else:
        question1 = df.loc[question1_index]
        question2 = df.loc[question2_index]
        embedding1 = model.encode(question1['GQ2'], convert_to_tensor=True)
        embedding2 = model.encode(question2['GQ2'], convert_to_tensor=True)
        cosine_similarity = util.pytorch_cos_sim(embedding1, embedding2)
        return cosine_similarity.item()


def get_random_question():
    global selected_questions
    # Ensure there are still questions left to select
    if len(df) == len(selected_questions):
        print("No more questions available.")
        return None

    # Select a random question that hasn't been selected yet
    while True:
        random_question = df.sample(n=1).iloc[0]
        # Use the index of the question as a hashable type
        question_index = random_question.name
        if question_index not in selected_questions:
            selected_questions.add(question_index)
            return random_question
```

```python
def get_next_question(current_question_index, preference):
    global selected_questions
    if preference == 'random':
        return get_random_question()
    elif preference == 'similar_bloom':
        # Assuming you have a way to find the index of the most similar question
        similar_question_index = find_most_similar_question_index(current_question_index, 'bloom')
        selected_questions.add(similar_question_index)
        return df.loc[similar_question_index]
    elif preference == 'similar_semantic':
        similar_question_index = find_most_similar_question_index(current_question_index, 'semantic')
        selected_questions.add(similar_question_index)
        return df.loc[similar_question_index]
```

```python
def find_most_similar_question_index(current_question_index, preference):
    global df
    if preference == 'bloom':
        similarity_scores = df.apply(lambda x: bloom_similarity(current_question_index, x.name), axis=1)
    elif preference == 'semantic':
        similarity_scores = df.apply(lambda x: semantic_similarity(current_question_index, x.name), axis=1)
    else:
        raise ValueError("Invalid preference type")

    # Find the index of the question with the highest similarity score
    most_similar_index = similarity_scores.idxmax()
    return most_similar_index
```

# LDA Clustering

```python
def preprocess_text2tokens(text):
    # Tokenization
    tokens = word_tokenize(text)
    # Remove stopwords and punctuation
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word.lower() not in stop_words and word not in string.punctuation]
    # Join filtered tokens back into a sentence
    token_text = ' '.join(filtered_tokens)
    return token_text
```

```python
df['MS Keywords'] = df['Mark Scheme'].apply(preprocess_text2tokens)
df['GQ Keywords'] = df['GQ2'].apply(preprocess_text2tokens)
df.head()
```

**Mark Scheme LDA**

```python
exam_questions_column = df['MS Keywords'].tolist()

# Tokenize the exam questions
tokenized_questions = [gensim.utils.simple_preprocess(question) for question in exam_questions_column]

# Create a dictionary representation of the documents
dictionary = corpora.Dictionary(tokenized_questions)

# Create a document-term matrix
doc_term_matrix = [dictionary.doc2bow(tokens) for tokens in tokenized_questions]

# Apply LDA model
num_topics = 5  # You can adjust the number of topics as needed
MS_lda_model = gensim.models.LdaModel(doc_term_matrix, num_topics=num_topics, id2word=dictionary, passes=20)

# Print the topics
pprint(MS_lda_model.print_topics())
```

```python
# Create a function to get the LDA topic category for each question
def get_topic_category(lda_model, dictionary, question):
    # Tokenize the question
    tokens = gensim.utils.simple_preprocess(question)
    # Convert the tokens into a bag of words
    bow = dictionary.doc2bow(tokens)
    # Get the LDA topic distribution for the question
    topic_distribution = lda_model.get_document_topics(bow)
    # Sort the topic distribution by probability and get the most probable topic
    topic_category = sorted(topic_distribution, key=lambda x: x[1], reverse=True)[0][0]
    return topic_category
```

```python
# Create a new column in the DataFrame to store the LDA topic category
df['MS_LDA'] = df['MS Keywords'].apply(lambda x: get_topic_category(MS_lda_model, dictionary, x))
df.head()
```
